Testing out ipython notebook integration with github pages

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline  
```

```python
df = pd.read_csv("ENB2012_data.csv")
df = pd.DataFrame(df, columns=['X6','Y1','Y2'])
df.describe()

# Note: Y1/Y2 is numerical data, X6 is categorical data.
# Use X6 to group by/create bins, 
# visualized Y1/Y2 in terms of bins made from X6
```

```python
plt.figure()

bp = df.boxplot(by='X6', showfliers = True, sym='k.')

plt.show()
```

```python
sns.boxplot(x="X6", y="Y1", data=df)
```

```python
sns.boxplot(x="X6", y="Y2", data=df)
```

```python
sns.boxplot(x="X6", y="Y1", data=df)
sns.stripplot(x="X6", y="Y1", data=df,
              size=4, jitter=True, edgecolor="gray")
```
